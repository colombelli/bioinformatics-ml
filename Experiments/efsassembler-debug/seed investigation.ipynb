```python
import pandas as pd
import numpy as np
import pickle

luad_path = "/home/colombelli/Documents/experiments/Experiments22_ago/luad/"
prad_path = "/home/colombelli/Documents/experiments/Experiments22_ago/prad/"
brca_path ="/home/colombelli/Documents/experiments/Experiments20_ago/"

hom_exps = ["hom_relieff", "hom_geode", "hom_gr", "hom_su", "hom_wx"]
sin_exps = ["sin_relieff", "sin_geode", "sin_gr", "sin_su", "sin_wx"]

exps = ["hyb_stbw_borda", "hyb_borda_borda", "het"] + hom_exps + sin_exps
```

```python
f = "/fold_sampling.pkl"

luad_exps = {}
for e in exps:
    file_complete_path = luad_path + e + f
    with open(file_complete_path, "rb") as input_file:
        luad_exps[e] = pickle.load(input_file)
```

```python
flag = 0
for key, val in luad_exps.items():
    for key2, val2 in luad_exps.items():
        print("\n\nComparing %s with %s:" % (key, key2))
        
        print("\nComparing test fold:")
        for i in range(len(val)):
            same = set(val[i][1]) == set(val[i][1])
            if not same:
                flag=1
            print(same)
            
        print("\nComparing traing folds:")
        for i in range(len(val)):
            same = set(val[i][0]) == set(val[i][0])
            if not same:
                flag=1
            print(same)
```

```python
flag
```
