```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def get_results_file(experiment):
    
    root_path = "/home/colombelli/Documents/Experiments09_abr/BRCA/"
    results_file_name = "/final_results.csv"
    complete_path = root_path + experiment + results_file_name
    return pd.read_csv(complete_path)
```

```python
hyb = get_results_file("Hyb_mean_mean")
het = get_results_file("Het_mean")
hom_geode = get_results_file("Hom_mean_geode")
hom_gr = get_results_file("Hom_mean_gr")
hom_su = get_results_file("Hom_mean_su")
hom_oner = get_results_file("Hom_mean_oner")
hom_relieff = get_results_file("Hom_mean_relieff")
```

```python
hyb
```

```python
hyb.plot(x='th_int', y='mean_ACC')
het.plot(x='th_int', y='mean_ACC')
```

```python
np.squeeze(hyb.loc[0:, ['mean_ACC']].values)
```

```python
x = np.squeeze(hyb.loc[0:, ['th_int']].values)
```

```python
def get_array_mean_acc(df):
    return np.squeeze(df.loc[0:, ['mean_ACC']].values)
```

```python
hyb_acc = get_array_mean_acc(hyb)
het_acc = get_array_mean_acc(het)
hom_geode_acc = get_array_mean_acc(hom_geode)
```

```python
plt.plot(x, hyb_acc, label = "Hybrid")
plt.plot(x, het_acc, label = "Heterogeneous")
plt.plot(x, hom_geode_acc, label = "Homogeneous-geoDE")

plt.xlabel('num genes')
plt.ylabel('accuracy')

plt.title('Hyb vs Het vs Hom-geoDE')
plt.legend()
plt.show()
```

```python
def get_array_stb(df):
    return np.squeeze(df.loc[0:, ['stabilities']].values)
```

```python
hyb_stb = get_array_stb(hyb)
het_stb = get_array_stb(het)
hom_geode_stb = get_array_stb(hom_geode)
```

```python
plt.plot(x, hyb_stb, label = "Hybrid")
plt.plot(x, het_stb, label = "Heterogeneous")
plt.plot(x, hom_geode_stb, label = "Homogeneous-geoDE")

plt.xlabel('num genes')
plt.ylabel('stability')

plt.title('Hyb vs Het vs Hom-geoDE')
plt.legend()
plt.show()
```

```python
experiments = [hyb, het, hom_geode, hom_gr, hom_su, hom_oner, hom_relieff]
labels = ["hyb", "het", "hom_geode", "hom_gr", "hom_su", "hom_oner", "hom_relieff"]


for i, exp in enumerate(experiments):
    stb = get_array_stb(exp)
    plt.plot(x, stb, label=labels[i])

plt.xlabel('num genes')
plt.ylabel('stability')

plt.title('Stabilities Comparsion')
plt.legend()
plt.show()
```

```python
experiments = [hyb, het, hom_geode, hom_gr, hom_su, hom_oner, hom_relieff]
labels = ["hyb", "het", "hom_geode", "hom_gr", "hom_su", "hom_oner", "hom_relieff"]


for i, exp in enumerate(experiments):
    acc = get_array_mean_acc(exp)
    plt.plot(x, acc, label=labels[i])

plt.xlabel('num genes')
plt.ylabel('accuracy')

plt.title('Accuracy Comparsion')
plt.legend()
plt.show()
```

```python
def get_array_mean_roc_auc(df):
    return np.squeeze(df.loc[0:, ['mean_ROC_AUC']].values)

def get_array_mean_pr_auc(df):
    return np.squeeze(df.loc[0:, ['mean_PR_AUC']].values)
```

```python
experiments = [hyb, het, hom_geode, hom_gr, hom_su, hom_oner, hom_relieff]
labels = ["hyb", "het", "hom_geode", "hom_gr", "hom_su", "hom_oner", "hom_relieff"]


for i, exp in enumerate(experiments):
    auc = get_array_mean_roc_auc(exp)
    plt.plot(x, auc, label=labels[i])

plt.xlabel('num genes')
plt.ylabel('roc auc')

plt.title('ROC-AUC Comparsion')
plt.legend()
plt.show()
```

```python
experiments = [hyb, het, hom_geode, hom_gr, hom_su, hom_oner, hom_relieff]
labels = ["hyb", "het", "hom_geode", "hom_gr", "hom_su", "hom_oner", "hom_relieff"]


for i, exp in enumerate(experiments):
    pr = get_array_mean_pr_auc(exp)
    plt.plot(x, pr, label=labels[i])

plt.xlabel('num genes')
plt.ylabel('precision-recall AUC')

plt.title('PR-AUC Comparsion')
plt.legend()
plt.show()
```

```python
experiments = [hyb, het, hom_geode, hom_gr, hom_su, hom_oner, hom_relieff]
labels = ["hyb", "het", "hom_geode", "hom_gr", "hom_su", "hom_oner", "hom_relieff"]


for i, exp in enumerate(experiments):
    stb = get_array_stb(exp)
    auc = get_array_mean_roc_auc(exp)
    tradeoff = (auc+1) * (stb+1)
    plt.plot(x, tradeoff, label=labels[i])

plt.xlabel('num genes')
plt.ylabel('trade-off')

plt.title('Trade-off Comparsion')
plt.legend()
plt.show()
```

```python
def get_results_file(experiment):
    
    root_path = "/home/colombelli/Documents/Experiments09_abr/KIRP/"
    results_file_name = "/final_results.csv"
    complete_path = root_path + experiment + results_file_name
    return pd.read_csv(complete_path)


hyb = get_results_file("Hyb_mean_mean")
het = get_results_file("Het_mean")
hom_geode = get_results_file("Hom_mean_geode")
hom_gr = get_results_file("Hom_mean_gr")
hom_su = get_results_file("Hom_mean_su")
hom_oner = get_results_file("Hom_mean_oner")
hom_relieff = get_results_file("Hom_mean_relieff")
```

```python
def plot_results(metric_foo, y_lab, title):
    
    experiments = [hyb, het, hom_geode, hom_gr, hom_su, hom_oner, hom_relieff]
    labels = ["hyb", "het", "hom_geode", "hom_gr", "hom_su", "hom_oner", "hom_relieff"]

    for i, exp in enumerate(experiments):
        result = metric_foo(exp)
        plt.plot(x, result, label=labels[i])

    plt.xlabel('num genes')
    plt.ylabel(y_lab)

    plt.title(title)
    plt.legend()
    plt.show()
```

```python
print("\n\tDATASET:   KIRP \n")
plot_results(get_array_mean_roc_auc, "roc auc", "ROC AUC Comparsion")
plot_results(get_array_mean_pr_auc, "pr auc", "PR AUC Comparsion")
plot_results(get_array_mean_acc, "accuracy", "Accuracy Comparsion")
plot_results(get_array_stb, "stability", "Stability Comparsion")
```
