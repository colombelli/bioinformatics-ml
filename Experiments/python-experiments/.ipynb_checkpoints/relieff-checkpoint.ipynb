```python
import numpy as np
import pandas as pd
import rpy2.robjects as robjects
from rpy2.robjects import pandas2ri
from rpy2.robjects.conversion import localconverter

file_path = dataset_path = "/home/colombelli/Documents/datasets/iqrSelectedGenesAllSamples.rds"
read_RDS = robjects.r['readRDS']
df = read_RDS(file_path)

with localconverter(robjects.default_converter + pandas2ri.converter):
    df = robjects.conversion.rpy2py(df)
```

```python
data = np.array(df.iloc[0:200, 0:len(df.columns)-1])
labels = np.array(df.iloc[0:200, len(df.columns)-1:]).flatten()
```

```python
from ReliefF import ReliefF

fs = ReliefF(n_neighbors=len(labels)-1)
X_train = fs.fit(data, labels)
```

```python
genes = list(df.columns)
for feature in fs.top_features:
    print(genes[feature])
```
