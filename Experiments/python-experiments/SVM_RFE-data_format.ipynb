```python
import pandas as pd
from sklearn.feature_selection import RFE
from sklearn.svm import SVR
```

```python
print("Reading dataset...")
df = pd.read_csv('~/Documents/datasets/testing.csv')


X = df.iloc[0:, 1:-1]
y = df.iloc[:, -1]


estimator = SVR(kernel="linear")
selector = RFE(estimator, 100, step=1)
print("Selecting the top 100 most relevant features...")
selector = selector.fit(X, y)
```

```python
df2 = pd.read_csv('~/Documents/datasets/testing.csv', index_col=0)
```

```python
X = df2.iloc[0:, 0:-1]
y = df2.iloc[:, -1]
```

```python
estimator2 = SVR(kernel="linear")
selector2 = RFE(estimator2, 1, step=1)
print("Selecting the top 100 most relevant features...")
selector22 = selector2.fit(X, y)
```

```python
print(selector22.support_)
```

```python
print(selector22.ranking_)
```

```python
data = {"rank": selector22.ranking_}
ranking = pd.DataFrame(data, index=list(X.columns))
```
