```python
import pandas as pd
import numpy as np
from math import log2
```

```python
file = "/home/colombelli/Documents/datasets/Projeto Bernardo/Agilent/GSE53752/Tabela dos dados normalizados GSE53752.txt"

mic_arr = pd.read_csv(file, sep=" ", )
mic_arr = mic_arr.set_index('GeneName')
del mic_arr['Description']
mic_arr = mic_arr.transpose().rename(columns={np.nan:'class'})
mic_arr["class"] = mic_arr["class"].replace({"Normal":0, "Tumor":1})
mic_arr
```

```python
rank = pd.read_csv("/home/colombelli/Documents/Experiments08_jul/BRCA/Het_borda/fold_1/agg_ranking_th200.csv")
genes = list(rank['Unnamed: 0'][0:200])

donthave = []
have = []
for gene in genes:
    try:
        mic_arr[gene]
        have.append(gene)
    except:
        donthave.append(gene)
```

```python
df = mic_arr[have + ['class']]
df = df.astype('float64').mean(axis=1, level=0)
df = df.dropna()
df
```

```python
main = pd.read_csv("/home/colombelli/Documents/datasets/assembler/brca_microarray.csv", index_col=0)
```

```python
df_train = main[have + ['class']]
df_train
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
from sklearn import metrics

def get_x(df):
    return df.loc[:, df.columns != 'class']
    
def get_y(df):
    return df.loc[:, ['class']].T.values[0]

def get_probs_positive_class(pred):
        positive_probs = []

        for prediction in pred:
            positive_probs.append(prediction[1])
        return positive_probs
```

```python
clf = SVC(gamma='auto', probability=True)

X_train = get_x(df_train)
y_train = get_y(df_train)

X_test = get_x(df)
y_test = get_y(df)


clf.fit(X_train, y_train)
clf.score(X_test, y_test)
#np.argmax(clf.predict_proba(X_test), axis=1)
```
