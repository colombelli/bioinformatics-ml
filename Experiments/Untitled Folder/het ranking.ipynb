```python
import pandas as pd
import numpy as np
from math import log2
```

```python
info = pd.read_csv("/home/colombelli/Documents/datasets/rna-seq/GSE75688_final_sample_information.txt", sep="\t")
df = pd.read_csv("/home/colombelli/Documents/datasets/rna-seq/GSE75688_GEO_processed_Breast_Cancer_raw_TPM_matrix.txt", sep="\t", index_col=1)
```

```python
del df['gene_id']
del df['gene_type']
```

```python
tpdf = df.transpose()
tpdf
```

```python
non_tumoral = []
tumoral = []

for row in info.index:
    
    for val in info.iloc[row][2:]:
        if val == 'Tumor':
            tumoral.append(info.iloc[row][0])
            break
        if val == 'nonTumor':
            non_tumoral.append(info.iloc[row][0])
            break
```

```python
from copy import deepcopy

df = deepcopy(tpdf)
```

```python
#clean non-defined indexes 
for i, sample in enumerate(df.index):
    if (not(sample in tumoral)) and (not(sample in non_tumoral)):
        df = df.drop(sample)
        
df
```

```python
to_drop = []
for i, sample in df.iterrows():
    
    if "Pooled" in i:
        to_drop.append(i)

df = df.drop(to_drop)
df
```

```python
df_log = (df+1).applymap(log2)

class_col = []
for sample in df_log.index:
    if sample in tumoral:
        class_col.append(1)
    elif sample in non_tumoral:
        class_col.append(0)
    else:
        print("bug on sample:", sample)

df_log['class'] = class_col
df_log
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
from sklearn import metrics


def get_x(df):
    return df.loc[:, df.columns != 'class']
    
def get_y(df):
    return df.loc[:, ['class']].T.values[0]

def get_probs_positive_class(pred):
    return [i[1] for i in pred]
```

```python
rank = pd.read_csv("/home/colombelli/Documents/Experiments08_jul/BRCA/rna-seq/Het_borda/fold_1/agg_ranking_th150.csv")
genes = list(rank['Unnamed: 0'][0:150])

donthave = []
have = []
for gene in genes:
    
    try:
        df_log[gene]
        have.append(gene)
    except:
        donthave.append(gene)
        
len(have)
```

```python
df_sel = df_log[have+['class']]
```

```python
X_train = get_x(df_sel)
y_train = get_y(df_sel)

clf = SVC(gamma='auto', probability=True)
clf.fit(X_train, y_train)

# accuracy
clf.score(X_train, y_train)
```

```python
pred = clf.predict_proba(X_train)
pred = get_probs_positive_class(pred)
metrics.roc_auc_score(np.array(y_train, dtype=int), pred)
```

```python
y_pred = np.argmax(clf.predict_proba(X_train), axis=1)
metrics.accuracy_score(y_train, y_pred)
```

```python
import rpy2.robjects as robjects
from rpy2.robjects import pandas2ri
from rpy2.robjects.conversion import localconverter
```

```python

def load_RDS(file_path):

    read_RDS = robjects.r['readRDS']
    return read_RDS(file_path)



def load_csv(file_path):
    return pd.read_csv(file_path, index_col=0)



def pandas_to_r(df):

    with localconverter(robjects.default_converter + pandas2ri.converter):
        r_from_pandas_df = robjects.conversion.py2rpy(df)
    return r_from_pandas_df



def r_to_pandas(df):
    with localconverter(robjects.default_converter + pandas2ri.converter):
            pandas_from_r_df = robjects.conversion.rpy2py(df)
    return pandas_from_r_df
```
