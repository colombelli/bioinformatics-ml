```python
import numpy as np
import pandas as pd
import math

import rpy2.robjects as robjects
from rpy2.robjects import pandas2ri
from rpy2.robjects.conversion import localconverter

from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
from sklearn import metrics


def get_x(df):
    return df.loc[:, df.columns != 'class']
    
def get_y(df):
    return df.loc[:, ['class']].T.values[0]


def load_RDS(file_path):

    read_RDS = robjects.r['readRDS']
    return read_RDS(file_path)



def load_csv(file_path):
    return pd.read_csv(file_path, index_col=0)



def pandas_to_r(df):

    with localconverter(robjects.default_converter + pandas2ri.converter):
        r_from_pandas_df = robjects.conversion.py2rpy(df)
    return r_from_pandas_df



def r_to_pandas(df):
    with localconverter(robjects.default_converter + pandas2ri.converter):
            pandas_from_r_df = robjects.conversion.rpy2py(df)
    return pandas_from_r_df
```

```python
mic_arr = pd.read_csv("/home/colombelli/Documents/datasets/Projeto Bernardo/Agilent/GSE70947/Tabela dos dados normalizados GSE70947.txt", sep=" ", )
```

```python
mic_arr = mic_arr.set_index('GeneName')
del mic_arr['Description']
mic_arr = mic_arr.transpose().rename(columns={np.nan:'class'})
mic_arr["class"] = mic_arr["class"].replace({"Normal":0, "Tumor":1})
mic_arr
```

```python
rank = pd.read_csv("/home/colombelli/Documents/Experiments11_mai/BRCA/Hyb_stb_borda/fold_1/agg_ranking_th50.csv")
genes = list(rank['Unnamed: 0'][0:50])

donthave = []
have = []
for gene in genes:
    
    try:
        mic_arr[gene]
        have.append(gene)
    except:
        donthave.append(gene)
        
len(have)
```

```python
df = mic_arr[have + ['class']]
```

```python
unique_genes = []
count=2
for i, gene in enumerate(df.columns):
    
    if not any(gene in string for string in unique_genes):
        unique_genes.append(gene)
        count=2
    else:
        gene_new = gene+"__"+str(count)
        
        if count==2:
            prev = unique_genes[-1]
            new = prev + "__1"
            unique_genes[-1] = new
        
        count+=1
        unique_genes.append(gene_new)
        
unique_genes
```

```python
df.columns = unique_genes
```

```python
from copy import deepcopy
curated_df = deepcopy(df)


def merge_columns(gene):
    
    columns_to_merge = []
    for column in curated_df.columns:
        if gene in column:
            columns_to_merge.append(deepcopy(curated_df[column]))
            del curated_df[column]
    
    summation = None
    count = 0
    for column in columns_to_merge:
        count+=1
        try:
            summation += column
        except: #first column
            summation = column
    
    mean = summation / count
    curated_df[gene] = mean
    return
```

```python
curated_genes = []
for i, gene in enumerate(df.columns):
    
    if '__' in gene: #then it has more than one
        radical = gene.split('__')[0]
        if radical in curated_genes:
            continue
        else:
            curated_genes.append(radical)
            merge_columns(radical)
```

```python
curated_df = pd.concat([curated_df, df.loc[:, 'class']], axis=1)
curated_df
```

```python
X_train = get_x(curated_df)
y_train = get_y(curated_df)

clf = SVC(gamma='auto', probability=True)
clf.fit(X_train, y_train)

# accuracy
clf.score(X_train, y_train)
```
