#### Data Structure

A list **[element1, element2, element3, element4 ...]** of tuples **(fst_element, snd_element)**

* Each tuple is one fold (train / test) data split;
* The tuple contains two lists:
    - 1st list: training samples for that fold iteration **[samp1, samp2, samp3]**
    - 2nd list: testing samples for that fold iteration **[samp4, samp5, samp6]**

### Simple functions to check if there is any repeated sample inside two lists

```python
def element_count(element, lists):
    
    count = 0
    for lst in lists:
        for list_el in lst:
            
            if list_el == element:
                count += 1
    return count
    
    

def count_duplicates(lst1, lst2):
    
    for element in lst1:
        
        count = element_count(element, [lst1, lst2])
        if  count != 1:
            print("List 1 have duplicates for element:", element)
            print("Number of repeating elements found:", count, "\n")
    
    
    for element in lst2:
        
        count = element_count(element, [lst1, lst2])
        if  count != 1:
            print("List 2 have duplicates for element:", element)
            print("Number of repeating elements found:", count, "\n")
        
```

### Building the SVM

```python
from sklearn.svm import SVC

clf = SVC(gamma='auto', probability=True)
clf.fit(training_x, training_y)
```

gamma{â€˜scaleâ€™, â€˜autoâ€™} or float, optional (default=â€™scaleâ€™)

    Kernel coefficient for â€˜rbfâ€™, â€˜polyâ€™ and â€˜sigmoidâ€™.

        if gamma='scale' (default) is passed then it uses 1 / (n_features * X.var()) as value of gamma,

        if â€˜autoâ€™, uses 1 / n_features.
        
        
        
------------------------------------------------------------------------------------------------------------------

probability: boolean, optional (default=False)

       Whether to enable probability estimates. This must be enabled prior to calling fit, will slow down that method as it internally uses 5-fold cross-validation, and predict_proba may be inconsistent with predict. 


```python
clf.score(testing_x, testing_y)
```

### Getting AUC values for the built SVM models

```python
from sklearn import metrics


def get_probs_positive_class(pred):
        positive_probs = []

        for prediction in pred:
            positive_probs.append(prediction[1])
        return positive_probs
```

```python
pred = clf.predict_proba(testing_x)
pred = get_probs_positive_class(pred)

print("First model")
print(metrics.roc_auc_score(np.array(testing_y, dtype=int), pred))


pred2 = selected_genes_clf.predict_proba(selected_genes_testing_x)
pred2 = get_probs_positive_class(pred2)

print("\n\nSecond model")
print(metrics.roc_auc_score(np.array(selected_genes_testing_y, dtype=int), pred2))

```

### Some extra informations about SVM's scikit-learn implementation

The **kernel** parameter is optional, if it is not provided, the algorithm will use rbf as its kernel.
The rbf kernel formula is given by: \exp(-\gamma \|x-x'\|^2)
More info: https://scikit-learn.org/stable/modules/svm.html#svm-kernels
