# Testing with another FS algorithms

```r
# OneR

att.scores2 <- oneR(chd ~ ., SAheart)
print(att.scores2)
print(cutoff.biggest.diff(att.scores2))
print(cutoff.k(att.scores2, k = 4))
print(cutoff.k.percent(att.scores2, 0.4))
```

```r
# Information Gain

att.scores3 <- information.gain(chd ~ ., SAheart)
print(att.scores3)
print(cutoff.biggest.diff(att.scores3))
print(cutoff.k(att.scores3, k = 4))
print(cutoff.k.percent(att.scores3, 0.4))
```

```r
# Gain Ratio

att.scores4 <- gain.ratio(chd ~ ., SAheart)
print(att.scores4)
print(cutoff.biggest.diff(att.scores4))
print(cutoff.k(att.scores4, k = 4))
print(cutoff.k.percent(att.scores4, 0.4))
```

```r
# Symmetrical Uncertainty

att.scores5 <- symmetrical.uncertainty(chd ~ ., SAheart)
print(att.scores5)
print(cutoff.biggest.diff(att.scores5))
print(cutoff.k(att.scores5, k = 4))
print(cutoff.k.percent(att.scores5, 0.4))
```
