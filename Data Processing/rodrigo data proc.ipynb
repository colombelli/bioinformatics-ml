```python
import pandas as pd
import numpy as np
from math import log2

file = "/home/colombelli/Documents/datasets/Projeto Bernardo/Agilent/GSE53752/Tabela dos dados normalizados GSE53752.txt"

mic_arr = pd.read_csv(file, sep=" ", )
mic_arr = mic_arr.set_index('GeneName')
del mic_arr['Description']
mic_arr = mic_arr.transpose().rename(columns={np.nan:'class'})
mic_arr["class"] = mic_arr["class"].replace({"Normal":0, "Tumor":1})
reord = list(mic_arr.columns[1:])
mic_arr = mic_arr[reord+["class"]]
mic_arr = mic_arr.astype('float64').mean(axis=1, level=0)
mic_arr = mic_arr.dropna()
mic_arr
```

```python
save_path = "/home/colombelli/Documents/datasets/rodrigo_proc/"
file_name = "GSE53752.csv"

mic_arr.to_csv(save_path+file_name)
pd.read_csv(save_path+file_name, index_col=0)
```

```python
mic_arr_log = (mic_arr.iloc[:, 0:-1]+1).applymap(log2)
mic_arr_log['class'] = np.array(mic_arr['class'], dtype='int')
mic_arr_log
```

```python
save_path = "/home/colombelli/Documents/datasets/rodrigo_proc/"
file_name = "GSE53752_log2.csv"

mic_arr_log.to_csv(save_path+file_name)
pd.read_csv(save_path+file_name, index_col=0)
```
