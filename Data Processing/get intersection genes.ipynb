```python
import pandas as pd
import numpy as np

cumida_path="/home/colombelli/Documents/datasets/cumida/full_proc/"
gses=["GSE22820", "GSE42568", "GSE70947"]

dfs = {}
for gse in gses:
    dfs[gse] = pd.read_csv(cumida_path+gse+"_npn.csv", index_col=0)
```

```python
tcga = "/home/colombelli/Documents/datasets/assembler/brca_rna_npn.csv"
dfs["tcga"] = pd.read_csv(tcga, index_col=0)
```

```python
len(dfs)
```

```python
genes = []

for key in dfs:
    g = list(dfs[key].columns)
    genes.append(g)
```

```python
genes_intersec = list(set.intersection(*[set(x) for x in genes])) 
```

```python
len(genes_intersec)
```

```python
no_cls = [x for x in genes_intersec if x != 'class']
```
