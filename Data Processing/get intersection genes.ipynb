### Intersection between microarrays brca data (tcga + cumida)

```python
import pandas as pd
import numpy as np

cumida_path="/home/colombelli/Documents/datasets/cumida/full_proc/"
gses=["GSE22820", "GSE42568", "GSE70947"]

dfs = {}
dfs['tcga'] = pd.read_csv("/home/colombelli/Documents/datasets/assembler/brca_microarray_no_drops.csv", index_col=0)
for gse in gses:
    dfs[gse] = pd.read_csv(cumida_path+gse+".csv", index_col=0)
```

```python
genes = []
for key in dfs:
    g = list(dfs[key].columns)
    genes.append(g)
    
genes_intersec = list(set.intersection(*[set(x) for x in genes]))
no_cls = [x for x in genes_intersec if x != 'class']
```

```python
base_path = "/home/colombelli/Documents/datasets/cumida/breast/intersect/"
for key in dfs:
    file_name = key+".csv"
    dfs[key].loc[:, no_cls+['class']].to_csv(base_path+file_name)
```
