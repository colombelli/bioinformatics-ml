```python
import seaborn as sns
import matplotlib.pyplot as plt

def boxplot_df(df):
    x = sns.boxplot(data=df)
    x.set(xticklabels=[])
    x.tick_params(bottom=False)
    x.set(xlabel=None)
    plt.show()
    return

def get_n_class(df, cls, n):
    selected = df.loc[df['class']==cls, :]
    selected = selected.sample(frac=1)
    try:
        return selected.iloc[:n, :]
    except:
        return selected
```

```python
"""

    Code to process ArrayExpression datasets
    
"""

def needs_translation(indexes):
    for col_name in indexes:
        if ("_at" in col_name) or ("AFFX-" in col_name):
            return True
    return False


def build_translation_dict(cols, gene_col, gpl):
    
    translation = {}
    for col in cols:
        print("Getting translation for", col)
        for i, val in enumerate(gpl[col]):
            
            if pd.isnull(val):
                continue
            
            if '//' in val:
                all_vals = val.split(' ')
                vals=[]
                for v in all_vals:
                    if '/' not in v:
                        vals.append(v)
                
            else: 
                vals=[val]

            for v in vals:
                
                gene_symbol = gpl[gene_col][i]
                if "//" in str(gene_symbol):
                    gene_symbol = gene_symbol.split(' ')[0]
                translation[v] = gene_symbol    
    return translation


def translation(df, gpl_file):
    
    if not needs_translation(df.columns):
        return df
    
    else:
        print("\nDataframe needs gene translation, loading gpl dataframe...")
        gpl = pd.read_csv(gpl_file, sep='\t', comment='#')
        
        gene_col = "Gene Symbol"
        cols = ["RefSeq Transcript ID",
                "ENTREZ_GENE_ID",
                "Representative Public ID",
                "ID",
                "GB_ACC"]
        translation_dict = build_translation_dict(cols, gene_col, gpl)
        
        counthave = 0
        countdont = 0
        for col in df.columns[1:]:
            if col in translation_dict.keys():
                counthave+=1
            else:
                countdont+=1
        print("counthave:", counthave, "\ncountdont", countdont)
        
        for col in df.columns:
            if not (col in translation_dict.keys()):
                translation_dict[col] = col
        
        cummulative = {}
        for col in df.columns:
            gene_symbol = translation_dict[col]
            try:
                cummulative[gene_symbol].append(df[col])
            except:
                cummulative[gene_symbol] = [df[col]]

        for key in cummulative:
            cummulative[key] = pd.concat(cummulative[key], axis=1).mean(axis=1).values
        
        proc = pd.DataFrame(cummulative)
        proc.index = df.index
        return proc

    
def build_class_col(df, class_dict):
    class_column = []
    missing_gsms = []
    for sample in df.index:
        try:
            try:
                class_column.append(class_dict[sample])  
            except:
                try: ## without .CEL
                    class_column.append(class_dict[sample.split('.')[0]])
                except: ## only the GSM, no _
                    class_column.append(class_dict[sample.split('_')[0]])
        except:
            missing_gsms.append(sample)
    
    if missing_gsms:
        print("\nThere are missing gsms class information: (", len(missing_gsms), ")")
        print(missing_gsms)
        print("\nDropping misses...")
        df = df.drop(index=missing_gsms)
        
    return class_column, df
                    


def process_data(df_file_path, relationship_df_file_path, saving_path, gpl_file=None):
    
    df = pd.read_csv(df_file_path, sep='\t', index_col=0).T
    relation = pd.read_csv(relationship_df_file_path, sep='\t')
    
    # infer if it needs translation and do it
    df = translation(df, gpl_file)
    
    # get class info
    class_dict= {}
    for i, row in relation.iterrows():
        if row[2] == "normal":
            sample_class = 0
        elif row[2] == "tumor":
            sample_class = 1
        else:
            sample_class = row[2]
            print("Unexpected class: ", row[2])
        class_dict[row[0]] = sample_class
        
    # build class column
    class_column, df = build_class_col(df, class_dict)            
    df['class'] = class_column
    
    # check if everything is fine with class info
    for i, value in enumerate(class_column):
        if (value != 0) and (value != 1):
            print("Generated class is neither 0 nor 1: (index, value)", i, value)
    
    # print class info
    print("\nClass information:")
    print(df['class'].value_counts())
    
    # save processed dataframe
    print("Saving processed dataframe in:", saving_path)
    df.to_csv(saving_path)
    
    # boxplots analysis
    print("\nNormal class only:")
    pldf = get_n_class(df, 0, 100).T
    boxplot_df(pldf)
    
    print("\nTumor class only:")
    pldf = get_n_class(df, 1, 100).T
    boxplot_df(pldf)
    
    print("\nBoth classes:")
    boxplot_df(pd.concat([get_n_class(df, 0, 50).T, get_n_class(df, 1, 50).T]))
    
    return df, relation
```

## Liver dataset processing

```python
df_file = "/home/colombelli/Documents/datasets/arrayexpress/liver/esets_liver_exprs.txt"
relationship_file = "/home/colombelli/Documents/datasets/arrayexpress/liver/E-MTAB-6695.sdrf.txt"
csv_file = "/home/colombelli/Documents/datasets/arrayexpress/liver/liver.csv"
gpl_file = "/home/colombelli/Documents/datasets/arrayexpress/liver/GPL570-55999.txt"
df, relation = process_data(df_file, relationship_file, csv_file, gpl_file)
```
