```python
import pandas as pd
import numpy as np

path = "/home/colombelli/Documents/datasets/cumida/"
```

```python
gse = "GSE22820/"

gpl = pd.read_table(path+gse+"gpl2.txt", sep='\t', comment='#')
gpl
```

```python
gpl = gpl[gpl['GENE_SYMBOL'].notna()].reset_index()
gpl
```

```python
transl = {}

for i, val in enumerate(gpl["ID"]):
    transl[val] = gpl["GENE_SYMBOL"][i]
    
print("id done")

for i, val in enumerate(gpl["SPOT_ID"]):
    transl[val] = gpl["GENE_SYMBOL"][i]

print("spot id done")
    
for i, val in enumerate(gpl["REFSEQ"]):
    transl[val] = gpl["GENE_SYMBOL"][i]

print("refeseq done")
    
for i, val in enumerate(gpl["ENSEMBL_ID"]):
    transl[val] = gpl["GENE_SYMBOL"][i]
    
print("done :D")
```

```python
for i, val in enumerate(gpl["ACCESSION_STRING"]):
    
    try:
        ac = val.split('|')
        for a in ac:
            transl[a] = gpl["GENE_SYMBOL"][i]
    except:
        pass
```

```python
gse = "GSE22820/"

gse = pd.read_csv(path+gse+"gse.csv", index_col=0)
gse
```

```python
counthave = 0
countdont = 0
for col in gse.columns[1:]:
    if col in transl.keys():
        counthave+=1
    else:
        countdont+=1

print("counthave:", counthave, "\ncountdont", countdont)
```

```python
gse['type'].unique()
```

```python
for col in gse.columns:

    if not (col in transl.keys()):
        transl[col] = col
```

```python
counthave = 0
countdont = 0
for col in gse.columns[1:]:
    if col in transl.keys():
        counthave+=1
    else:
        countdont+=1

print("counthave:", counthave, "\ncountdont", countdont)
```

```python
from copy import deepcopy

gse['type'] = gse['type'].replace('primary_breast_cancer', 1).replace('normal', 0)
transl['type'] = 'class'

df = deepcopy(gse)
```

```python
cummulative = {}

for col in df.columns:
    
    gene_symbol = transl[col]
    try:
        cummulative[gene_symbol].append(df[col])
    except:
        cummulative[gene_symbol] = [df[col]]
```

```python
for key in cummulative:
    #if len(cummulative[key]) > 1:
    cummulative[key] = pd.concat(cummulative[key], axis=1).mean(axis=1).values
```

```python
proc = pd.DataFrame(cummulative)
proc.index = df.index
proc
```

```python
proc.to_csv(path+"full_proc/GSE22820.csv")
```

# GSE42568

```python
gse = "GSE42568/"

gpl = pd.read_table(path+gse+"gpl.txt", sep='\t', comment='#')
gpl
```

```python
gpl = gpl[gpl['Gene Symbol'].notna()].reset_index()
gpl
```

```python
def build_translation_dict(cols, gene_col):
    
    translation = {}
    for col in cols:
        print("Getting translation for", col)
        for i, val in enumerate(gpl[col]):
            
            if pd.isnull(val):
                continue
            
            if '//' in val:
                all_vals = val.split(' ')
                vals=[]
                for v in all_vals:
                    if '/' not in v:
                        vals.append(v)
                
            else: 
                vals=[val]

            for v in vals:
                
                gene_symbol = gpl[gene_col][i]
                if "//" in gene_symbol:
                    gene_symbol = gene_symbol.split(' ')[0]
                translation[v] = gene_symbol
                
    return translation
    
```

```python
gene_col = "Gene Symbol"

cols = [
    "RefSeq Transcript ID",
    "ENTREZ_GENE_ID",
    "Representative Public ID",
    "ID",
    "GB_ACC",
    "SPOT_ID"
]

translation = build_translation_dict(cols, gene_col)
translation
```

```python
gse = pd.read_csv(path+gse+"gse.csv", index_col=0)
gse
```

```python
counthave = 0
countdont = 0
for col in gse.columns[1:]:
    if col in translation.keys():
        counthave+=1
    else:
        countdont+=1

print("counthave:", counthave, "\ncountdont", countdont)
```

# GSE70947

```python
gse = "GSE70947/"

gpl = pd.read_table(path+gse+"gpl.txt", sep='\t', comment='#')
gpl
```

```python
gpl = gpl[gpl['GeneName'].notna()].reset_index()
gpl
```

```python
def build_translation_dict(cols, gene_col):
    
    translation = {}
    for col in cols:
        print("Getting translation for", col)
        for i, val in enumerate(gpl[col]):
            
            if pd.isnull(val):
                continue
            
            if '|' in val:
                all_vals = val.split('|')
                vals=[]
                for v in all_vals:
                    if v and ('ref' not in v) and ('ens' not in v):
                        vals.append(v)
                
            else: 
                vals=[val]

            for v in vals: 
                gene_symbol = gpl[gene_col][i]
                translation[v] = gene_symbol
                
    return translation
    
```

```python
gene_col = "GeneName"

cols = [
    "ProbeName",
    "GB_ACC",
    "accessions"
]

translation = build_translation_dict(cols, gene_col)
translation
```

```python
counthave = 0
countdont = 0
for col in gse.columns[1:]:
    if col in translation.keys():
        counthave+=1
    else:
        countdont+=1

print("counthave:", counthave, "\ncountdont", countdont)
```
