```python
import pandas as pd
```

```python
file = "esets_breast_exprs_genes.txt"
fdir = "/home/colombelli/Documents/datasets/arrayexpress/breast/"
```

```python
df = pd.read_csv(fdir+file, sep='\t', index_col=0)
df
```

```python
relation = pd.read_csv(fdir+"E-MTAB-6703.sdrf.txt", sep='\t')
relation
```

```python
class_dict= {}
for i, row in relation.iterrows():
    if row[2] == "normal":
        sample_class = 0
    elif row[2] == "tumor":
        sample_class = 1
    else: 
        print("Unexpected class: ", row[2])
    
    class_dict[row[0]] = sample_class
```

```python
df = df.T
df
```

```python
class_column = []
for sample in df.index:
    class_column.append(class_dict[sample])
```

```python
for i, value in enumerate(class_column):
    if (value != 0) and (value != 1):
        print(i, value)
```

```python
df['class'] = class_column
df
```

```python
df.to_csv(fdir+"breast.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def boxplot_df(df):
    x = sns.boxplot(data=df)
    x.set(xticklabels=[])
    x.tick_params(bottom=False)
    x.set(xlabel=None)
    plt.show()
    return

def get_n_class(df, cls, n):
    selected = df.loc[df['class']==cls, :]
    selected = selected.sample(frac=1)
    try:
        return selected.iloc[:n, :]
    except:
        return selected
```

```python
#BOXPLOT NORMAL
pldf = get_n_class(df, 0, 100).T
boxplot_df(pldf)
```

```python
#BOXPLOT TUMOR
pldf = get_n_class(df, 1, 100).T
boxplot_df(pldf)
```

```python
#BOXPLOT NORMAL+TUMOR
boxplot_df(pd.concat([get_n_class(df, 0, 50).T, get_n_class(df, 1, 50).T]))
```
