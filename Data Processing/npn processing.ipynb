```r
library(huge)
basedir <- "/home/colombelli/Documents/datasets/assembler/"
rnaseq.file <- "brca_rna_symb_no_log.csv"
df <- read.csv(paste(basedir, rnaseq.file, sep=""), header=T, row.names=1,sep=",")
```

```r
df
```

```r
cls.cols <- df$class
df$class <- NULL
```

```r
npn.df <- huge.npn(df, npn.func = "shrinkage", npn.thresh = NULL, verbose = TRUE)
npn.df
```

```r
npn.df <- as.data.frame(npn.df)
```

```r
npn.df$class <- cls.cols
```

```r
npn.df
```

```r
save.file <- "brca_rna_npn.csv"
savedir <- paste(basedir, save.file, sep="")
savedir
```

```r
write.csv(npn.df, file = savedir, row.names = TRUE)
```
