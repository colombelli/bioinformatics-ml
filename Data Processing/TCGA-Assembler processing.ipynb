```python
import pandas as pd
import numpy as np

path = "/home/colombelli/Documents/TCGA-Assembler-2/TCGA-Assembler/Data/"
arr_file = "BRCA__gene_Array__tissueTypeAll__20200708161355.txt"
rna_file = "BRCA__gene.normalized_RNAseq__tissueTypeAll__20200708162118.txt"
```

```python
data = pd.read_csv(path+arr_file, sep="\t")
data = data.T
data.columns = data.iloc[0]
data = data.drop(data.index[0])
data
```

```python
classes = []

for barcode in data.index:
    sample_type = int(barcode.split('-')[3][0:-1])
    if sample_type > 9 and sample_type < 20:    # Codes within [10, 19] are normal sample types
        classes.append(0)
    elif sample_type > 0 and sample_type < 10:  # Codes within [1, 9] are tumoral sample types
        classes.append(1)
    else:
        print("\nControl sample type found for barcode:", barcode)
        
    if sample_type != 1 and sample_type != 6 and sample_type != 11:
        print("\nUnexpected sample type found for barcode:", barcode)
```

```python
np_cls = np.array(classes)
np.unique(np_cls, return_counts=True)
```

```python
len(np_cls)
```

```python
data['class'] = classes
```

```python
# RNA SEQ

data = pd.read_csv(path+rna_file, sep="\t")
data = data.T
data.columns = data.iloc[0]
data = data.drop(data.index[0])
data
```

```python
data.dropna()
```

```python
new_cols = []
count_no_gene_sym = 0

for gene in data.columns:
    symbol = gene.split('|')[0]
    if symbol == '?':
        new_cols.append(gene)
        count_no_gene_sym +=1
    else:
        new_cols.append(symbol)
```

```python
count_no_gene_sym
```

```python
data.columns = new_cols
data
```
