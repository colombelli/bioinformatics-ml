```python
import pandas as pd
import seaborn as sns

def boxplot_df(df):
    x = sns.boxplot(data=df)
    x.set(xticklabels=[])
    x.tick_params(bottom=False)
    x.set(xlabel=None)
    return
```

```python
rnaseq = "/home/colombelli/Documents/datasets/brca_rnaseq_genes_intersec.csv"
micarr = "/home/colombelli/Documents/datasets/assembler/brca_micarr_npn.csv"
df1_path = "/home/colombelli/Documents/datasets/cumida/intersect_genes/GSE42568.csv"
df2_path = "/home/colombelli/Documents/datasets/cumida/intersect_genes/GSE22820.csv"
df3_path = "/home/colombelli/Documents/datasets/cumida/intersect_genes/GSE70947.csv"
```

```python
gses = [df1_path, micarr]

dfs = {}
for path in gses:
    gse = path.split('/')[-2]
    dfs[gse] = pd.read_csv(path, index_col=0)

genes = []
for key in dfs:
    g = list(dfs[key].columns)
    genes.append(g)
    
genes_intersec = list(set.intersection(*[set(x) for x in genes])) 
no_cls = [x for x in genes_intersec if x != 'class']
print(len(no_cls))
```

```python
len(genes_intersec)
```

```python
df0 = pd.read_csv(rnaseq, index_col=0)
df1 = pd.read_csv(micarr, index_col=0)
df2 = pd.read_csv(df1_path, index_col=0)
df3 = pd.read_csv(df2_path, index_col=0)
df4 = pd.read_csv(df3_path, index_col=0)
```

```python
df0 = df0.loc[:, genes_intersec]
df1 = df1.loc[:, genes_intersec]
df2 = df2.loc[:, genes_intersec]
df3 = df3.loc[:, genes_intersec]
df4 = df4.loc[:, genes_intersec]
```

```python
dfs = [df0,df1,df2,df3,df4]
```

```python
dfs20 = []
for df in dfs:
    df = df.sample(frac=1)
    dfs20.append(df.iloc[0:20, 0:-1])
```

```python
dfConc = pd.concat([dfs20[1], dfs20[2]]).T
boxplot_df(dfConc)
```

```python
dfConc = pd.concat([dfs20[0], dfs20[1]]).T
boxplot_df(dfConc)
```

```python
dfConc = pd.concat(dfs20).T
boxplot_df(dfConc)
```
