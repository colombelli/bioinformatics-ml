```python
import pandas as pd
import numpy as np
from copy import deepcopy

lung_path = "/home/colombelli/Documents/datasets/cumida/lung/"
prostate_path = "/home/colombelli/Documents/datasets/cumida/prostate/"
```

```python
gse = "GSE27262/"

df = pd.read_csv(lung_path+gse+"gse.csv", index_col=0)
gpl = pd.read_csv(lung_path+gse+"gpl", sep='\t', comment='#')
gpl
```

```python
gpl = gpl[gpl['Gene Symbol'].notna()].reset_index()
gpl
```

```python
def build_translation_dict(cols, gene_col):
    
    translation = {}
    for col in cols:
        print("Getting translation for", col)
        for i, val in enumerate(gpl[col]):
            
            if pd.isnull(val):
                continue
            
            if '//' in val:
                all_vals = val.split(' ')
                vals=[]
                for v in all_vals:
                    if '/' not in v:
                        vals.append(v)
                
            else: 
                vals=[val]

            for v in vals:
                
                gene_symbol = gpl[gene_col][i]
                if "//" in gene_symbol:
                    gene_symbol = gene_symbol.split(' ')[0]
                translation[v] = gene_symbol
                
    return translation
```

```python
gene_col = "Gene Symbol"

cols = [
    "RefSeq Transcript ID",
    "ENTREZ_GENE_ID",
    "Representative Public ID",
    "ID",
    "GB_ACC",
    "SPOT_ID"
]

translation = build_translation_dict(cols, gene_col)
```

```python
counthave = 0
countdont = 0
for col in df.columns[1:]:
    if col in translation.keys():
        counthave+=1
    else:
        countdont+=1

print("counthave:", counthave, "\ncountdont", countdont)
for col in df.columns:

    if not (col in translation.keys()):
        translation[col] = col
```

```python
df['type'].unique()
```

```python
df['type'] = df['type'].replace('Non_small_cell_lung_cancer', 1).replace('normal', 0)
translation['type'] = 'class'
```

```python
cummulative = {}
for col in df.columns:
    gene_symbol = translation[col]
    try:
        cummulative[gene_symbol].append(df[col])
    except:
        cummulative[gene_symbol] = [df[col]]
        
for key in cummulative:
    cummulative[key] = pd.concat(cummulative[key], axis=1).mean(axis=1).values
```

```python
proc = pd.DataFrame(cummulative)
proc.index = df.index
proc
```

```python
proc.to_csv(lung_path+gse+"translated.csv")
```
